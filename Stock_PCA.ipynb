```python
import tensorflow as tf
import numpy as np
```

```python
#load stock_data(kospi) column means Open,High,Low,Close and Volume
data = np.genfromtxt('kospi_.csv',delimiter=',')
for i in range(10):
    print(data[i])
    #show datas from 0 row to 10 row
```

```python
#normalize data
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    # noise term prevents the zero division
    return numerator / (denominator+1e-10)
data_scaled=MinMaxScaler(data)
for i in range(10):
    print(data_scaled[i])
    #show datas from 0 row to 10 row
```

```python
#get mean and covariance vector like siraj's
mean_vec = np.mean(data_scaled, axis=0)
cov_mat = (data_scaled - mean_vec).T.dot((data_scaled - mean_vec)) / (data_scaled.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
print('NumPy covariance matrix: \n%s' %np.cov(data_scaled.T))
```

```python
#get eigenvector and eigenvalue
#check out our value is correct like siraj's
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Everything ok!')
```

```python
#check the eigenvalue order 
#show portion of eigenvalue to determine How many dimention would be reducted
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
    
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(5), var_exp, alpha=0.5, align='center',label='individual explained variance')
    plt.step(range(5), cum_var_exp, where='mid',label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.savefig('./pca2.pdf')
```

```python
#the first and second portions is big and the rest of them is not.
#PCA dimention is 2
#make matrix for PCA
matrix_w = np.hstack((eig_pairs[0][1].reshape(5,1),eig_pairs[1][1].reshape(5,1)))
print('Matrix W:\n', matrix_w)
```

```python
#the result of matrix should be analyzed
#the first column is [-0.5, -0.5, -0.47, -0.5, -0.01]. The 1,2,3,4 rows of 1st column are major portion and the value is similar
#the 2nd column is [0.006, 0.003, -0.017, -0.01, 0.99]. the 5 row of 2nd column is major(0.99)
#above 2 analysis means first dimention of PCA is determined by 'Open','High','Low','Close', not 'Volumn',
#and second dimention of PCA is determined by 'Volume'.
Y = data_scaled.dot(matrix_w)
with plt.style.context('seaborn-whitegrid'):
    fig = plt.figure()
    plt.hold(True)
    ax = plt.subplot(121)
    for i in range(4):
        plt.plot(data_scaled[:, i])
    plt.plot((-Y[:,0]))
    
    ax = plt.subplot(122)
    plt.plot(data_scaled[:,i+1])
    plt.plot(Y[:, 1])
    plt.hold(False)
    plt.savefig('./pca1.pdf')
```
